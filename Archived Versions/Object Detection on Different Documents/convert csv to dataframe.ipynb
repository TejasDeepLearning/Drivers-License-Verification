```python
import requests
import pandas as pd

urls = pd.read_csv('licenses.csv')
rows = []

for index, i in urls.iterrows():
    rows.append(i[-1])
    
counter = 0

for i in rows:
    file_name = 'license' + str(counter) + '.jpg'
    
    print(file_name)
    response = requests.get(i)
    file = open(file_name, 'wb')
    file.write(response.content)
    file.close()
    counter += 1
```
